# T-Test

**Definition:** T-Test is used to compare means when the population variance is unknown or the sample size is small.

---

## Independent T-Test

**Example:** Compare **Annual Income** of **Male vs Female** customers.  

**Hypotheses:**  
- H0: Distribution of tip for yellow = Distribution of tip for green
- H1:Distribution of tip for yellow â‰  Distribution of tip for green

```python
from scipy import stats
import statsmodels.api as sm
from statsmodels.stats.weightstats import ztest
from statsmodels.stats.multicomp import pairwise_tukeyhsd
```

```python
# Independent T-Test
yellow_taxi = df[df['color']=='yellow']['tip']
green_taxi = df[df['color']=='green']['tip']

t_stat, p_val = stats.ttest_ind(yellow_taxi, green_taxi)
print("Independent T-Test statistic:", t_stat)
print("P-value:", p_val)

if p_val < 0.05:
    print("Reject null hypothesis: tip for yellow taxi is different that of green")
else:
    print("Fail to reject null hypothesis: No significant tip difference by color")
```

Bar plot indicates that the average tip received when customers pay by 'credit card' is substantially higher than when they pay by 'cash'.

The histogram, with its KDE, visually confirms these observations. The distribution for 'credit card' tips is shifted towards higher values and has a longer tail, indicating more frequent larger tips. The 'cash' tip distribution is heavily concentrated at very low tip amounts, with a sharp drop-off, suggesting that cash payers either tip very little or not at all, or the tip is often not recorded.


**Hypotheses:**  
- H0: Distribution of tip by credit card = Distribution of tip by cash
- H1:Distribution of tip  by credit cardâ‰  Distribution of tip by cash

```python
# Independent T-Test
credit_card = df[df['payment']=='credit card']['tip']
cash= df[df['payment']=='cash']['tip']

t_stat, p_val = stats.ttest_ind(credit_card, cash)
print("Independent T-Test statistic:", t_stat)
print("P-value:", p_val)

if p_val < 0.05:
    print("Reject null hypothesis: tip from credit card is different that of cash")
else:
    print("Fail to reject null hypothesis: No significant tip difference by payment")
```

**Hypotheses:**  
- H0: distribution of tip is the same across all pickup_borough categories  
- H1: At least one category has a different distribution of tip

```python
manhattan_tip=df['tip'][df['pickup_borough']=='Manhattan']
queens_tip=df['tip'][df['pickup_borough']=='Queens']
bronx_tip=df['tip'][df['pickup_borough']=='Bronx']
brooklyn_tip=df['tip'][df['pickup_borough']=='Brooklyn']
```

```python
# ANOVA Test
f_stat, p_val = stats.f_oneway(manhattan_tip,queens_tip,bronx_tip,brooklyn_tip)
print("ANOVA F-statistic:", f_stat)
print("P-value:", p_val)

if p_val < 0.05:
    print("Reject null hypothesis: Spending Score differs by Region")
else:
    print("Fail to reject null hypothesis: No significant difference by Region")
```


**Hypotheses for Tukey test (for each pair of groups):**  
- H0:distribution of group1 = distribution of group2  
- H1: distribution of group1 â‰  distribution of group2

```python
# Tukey HSD Test
from statsmodels.stats.multicomp import pairwise_tukeyhsd

# Perform Tukey HSD
tukey_result = pairwise_tukeyhsd(endog=df.tip, groups=df.pickup_borough, alpha=0.05)
print(tukey_result)
```

H0:Corelation is zero

H1:there is significant corelation

```python
r,p_val=stats.spearmanr(df['tip'],df['passengers'])
print("correlation:",r)
print("p-value:",p_val)
alpha=0.05
if p_val<alpha:
  print("Reject null hypothesis:there is no relationship")
else:
  print("Fail to reject null hypothesis:there is significant relationship")
```
